```python
import xmltodict
import pprint
import numpy as np
import pandas as pd
import openpyxl
import os #Manipular pasta e arquivos do computador
```

```python
def ler_xml_danfe(nota):
    
    with open(nota, "rb") as arquivo:
        documento = xmltodict.parse(arquivo) #Converte XML para Dicionário 
    dic_notafiscal = documento['nfeProc']['NFe']['infNFe']
    #pprint.pprint(dic_notafiscal)
    valor_total = dic_notafiscal['total']['ICMSTot']['vNF']
    cnpj_venda = dic_notafiscal['emit']['CNPJ']
    nome_vendeu = dic_notafiscal['emit']['xNome']
    cpf_compra = dic_notafiscal['dest']['CPF']
    nome_comprou = dic_notafiscal['dest']['xNome']

    produtos = dic_notafiscal['det']
    lista_produtos = []

    for produto in produtos:
        valor_produto = produto['prod']['vProd']
        nome_produto = produto['prod']['xProd']
        lista_produtos.append((nome_produto, valor_produto))


    resposta = {
            'valor_total': [valor_total], 
            'cnpj_venda': [cnpj_venda], 
            'nome_vendeu': [nome_vendeu], 
            'cpf_compra': [cpf_compra], 
            'nome_comprou': [nome_comprou], 
            'lista_produtos': [lista_produtos]
           }
    
    return resposta

def ler_xml_servico(nota):
    
    with open(nota, 'rb') as arquivo:
        documento = xmltodict.parse(arquivo)
    #print(documento)
    dic_notafiscal = documento['ConsultarNfseResposta']['ListaNfse']['CompNfse']['Nfse']['InfNfse']
    
    valor_total = dic_notafiscal['Servico']['Valores']['ValorServicos']
    cnpj_vendeu = dic_notafiscal['PrestadorServico']['IdentificacaoPrestador']['Cnpj']
    nome_vendeu = dic_notafiscal['PrestadorServico']['RazaoSocial']
    cpf_comprou = dic_notafiscal['TomadorServico']['IdentificacaoTomador']['CpfCnpj']['Cnpj']
    nome_comprou = dic_notafiscal['TomadorServico']['RazaoSocial']
    produtos = dic_notafiscal['Servico']['Discriminacao']
    
    resposta = {
            'valor_total': [valor_total], 
            'cnpj_vendeu': [cnpj_vendeu], 
            'nome_vendeu': [nome_vendeu], 
            'cpf_comprou': [cpf_comprou], 
            'nome_comprou': [nome_comprou], 
            'lista_produtos': [produtos]        
    }
    
    return resposta

#pprint.pprint(ler_xml_danfe('DANFEBrota.xml'))
#pprint.pprint(ler_xml_danfe('DANFENespresso.xml'))

#Percorrer a pasta e fazer a leitura dos arquivos XML automaticamente

lista_arquivos = os.listdir(r'C:\Users\adria\Desktop\Projeto 2.1 - Leitura de Nota Fiscal (XML)')

for arquivo in lista_arquivos:
    if '.xml' in arquivo:
        #print(arquivo)
        if 'DANFE' in arquivo:
            print(ler_xml_danfe(arquivo))
        else:
            print(ler_xml_servico(arquivo))



#Salvar informações em um arquivo excel ou csv

#tabela = pd.DataFrame.from_dict(resposta)
#tabela.to_excel('NFs.xlsx')
        
#Caso queira adicionar todas NF's no mesmo excel

#arquivos = os.listdir(r'C:\Users\adria\Desktop\Projeto 2.1 - Leitura de Nota Fiscal (XML)')

#df_final = pd.DataFrame()
#for arquivo in arquivos:
    #if '.xml' in arquivo:
        #print(arquivo)
        #if 'DANFE' in arquivo:
            #df = pd.DataFrame.from_dict(ler_xml_danfe(arquivo))
        #else:
            #df = pd.DataFrame.from_dict(ler_xml_servico(arquivo))
            
        #df_final = df_final.append(df)
        
#print(df_final)
#df_final.to_excel('NFs.xlsx', index=False)


    



```
